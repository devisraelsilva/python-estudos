```python
#Importando base de dados

import pandas

#abrir a base de dados
diretorio_download_base = "/home/raels97/Downloads/Vendas - Dez.xlsx"
base_dados_vendas = pandas.read_excel(diretorio_download_base)

#visualizar base de dados
display(base_dados_vendas)

#somar faturamento de todos os produtos = coluna "Valor Final"
faturamento = base_dados_vendas["Valor Final"].sum()

#somar quantidade de produtos vendidos = coluna "Quantidade"
quantidade_produtos = base_dados_vendas["Quantidade"].sum()


print(faturamento)
print(quantidade_produtos)
```
