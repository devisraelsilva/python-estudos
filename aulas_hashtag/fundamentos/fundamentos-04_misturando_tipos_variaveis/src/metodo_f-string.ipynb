Variável está em tipo: int

Informação do print( ) é texto str/string, vou considerar como str para somar as informações, texto com números

```python
faturamento = 1500
custo = 500
lucro = faturamento - custo
#type(faturamento)

print("O faturamento da loja foi: " + str(faturamento) + " reais." + " O custo da loja foi: " + str(custo) + " reais." +" O lucro foi de: " + str(lucro) + " reais.")
```

Método F-STRING

Para utilizar f-string, adicionandono inicio entre ( ) um "f", indicando que alí haverão formatações e declarações de variáveis. Essa ferramenta possibilita ao usuário a criação de operações com as variáveis, como soma, subtração e etc.

```python
faturamento = 1500
custo = 500
lucro = faturamento - custo
#type(faturamento)

#puxando a variável direto:
print(f"O faturamento da loja foi: {faturamento} reais. O custo da loja foi: {custo} reais. O lucro foi de: {lucro} reais.")

#realizando um calculo para trazer o valor de lucro, ao invés de puxar direto:
print(f"O faturamento da loja foi: {faturamento} reais. O custo da loja foi: {custo} reais. O lucro foi de: {faturamento - custo} reais.")
```
