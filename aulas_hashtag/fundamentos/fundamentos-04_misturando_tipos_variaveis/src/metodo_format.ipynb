Variável está em tipo: int

Informação do print( ) é texto str/string, vou considerar como str para somar as informações, texto com números

```python
faturamento = 1500
custo = 500
lucro = faturamento - custo
#type(faturamento)

print("O faturamento da loja foi: " + str(faturamento) + " reais." + " O custo da loja foi: " + str(custo) + " reais." +" O lucro foi de: " + str(lucro) + " reais.")
```

Método FORMAT

Faço a definição do texto que quero exibir, onde aparecerá o valor da variável adiciono { }, depois declaro: ".format( )" entre parenteses qual variável retornará as informações. Mais de uma variável, separar por virgulas.

Atenção: declarar a variável na ordem correta de qual vai em cada { }, dentro do mesmo ( ).

```python
faturamento = 1500
custo = 500
lucro = faturamento - custo
#type(faturamento)

print("O faturamento da loja foi: {} reais. O custo da loja foi: {} reais. O lucro foi de: {} reais.".format(faturamento, custo, lucro))
```
