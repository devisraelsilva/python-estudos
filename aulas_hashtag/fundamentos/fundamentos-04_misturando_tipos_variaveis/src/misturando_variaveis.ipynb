```python
#variável está em tipo: float
#informação do print() é numero int, vou considerar como float para somar as informações e trazer decimais

faturamento = 1000.79
#type(faturamento)

print(150 + float(faturamento))
```

```python
#variável está em tipo: float
#informação do print() é numero int, vou considerar como int para somar as informações,
#porém não traz decimais

faturamento = 1000.79
#type(faturamento)

print(150 + int(faturamento))
```

```python
#variável está em tipo: int
#informação do print() é texto str/string, vou considerar como str para somar as informações,
#texto com números

faturamento = 1500
custo = 500
lucro = faturamento - custo
#type(faturamento)

print("O faturamento da loja foi: " + str(faturamento) + " reais")
print("O custo da loja foi: " + str(custo) + " reais")
print("O lucro foi de: " + str(lucro) + " reais")
```
