# Mudança do tipo de variável

Input - o usuário irá imputar os dados que serão atribuídos as variáveis.
Todo input realizado pelo usuário, automaticamente recebe as informações como TEXTO/STRING.

### str( ) -> muda para string
### int( ) -> muda para inteiro
### float( ) -> muda para float

```python
faturamento = float(input("Insira o valor do faturamento: "))
custo = float(input("Insira o valor do custo: "))

#print(type(faturamento)) # type( ) mostra qual o tipo da variável
#print(type(custo)) # type( ) mostra qual o tipo da variável

lucro = faturamento - custo

print(lucro)
```
